```python
from IPython.core.display import TextDisplayObject
import pandas as pd
import matplotlib.pyplot as mp
import numpy as np
import csv
from pathlib import Path
from matplotlib.backends.backend_pdf import PdfPages
from pathlib import Path
import os
import math
```

```python
# Setting filepath to all data folders
%store -r fp_folder
fp_targets = fp_folder+'Target_Data/'
```

```python
def GraphMovement(block):
  target_list = ['Automobile shop', 'Police station ', 'Fire Station', 'Bank', 'Pawn Shop', 'Pizzeria', 'Quattroki Restaurant', 'High School']
  for target in target_list:
    fp_file = fp_targets + block +'_'+target+'_results.csv'
    df = pd.read_csv(fp_file)
    df = df[:-1]
    mp.title(target + ' ' + block)
    #ax = mp.axes()
    df.set_index('X', inplace = True)
    #ax.set_xlabel('x position in maze')
    #ax.set_ylabel('y position in maze')
    ax = df.groupby(['Participant','Block_num'])['Z'].plot(legend=False, xlim = [-80, 80], ylim = [-60,80])
    imagepath = Path(fp_folder + '/Target_Data/maps/'+block+'/'+block+'_'+target+'.png') 
    imagepath.parent.mkdir(parents=False, exist_ok=True)
    mp.savefig(imagepath)
```

```python
os.makedirs(fp_targets+'maps/all/', exist_ok=True)
GraphMovement('all')
```
